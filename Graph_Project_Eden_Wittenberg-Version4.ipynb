```python
#Part 2:The Graph class
    #Task 1: Define the class
class graph:
    def __init__(self, name, nodes=[]):
        self.name=name
        self.nodes={}
        for n in nodes:
            self.nodes[n.name]=n

    def __str__(self):
        report='Graph name: {} | Number of nodes: {}'.format(self.name,len(self.nodes))+'\n'
        for nd_name, nd_object in self.nodes.items():
            report=report+ str(nd_object.__str__())+'\n'
        return report

    def __len__(self):
        return len(self.nodes)

    def __contains__(self, key):
        if (isinstance(key, str)):
            if (key in self.nodes.keys()):
                return True
            return False
        elif (isinstance(key, node)):
            if (key.name in self.nodes.keys()):
                return True
        return False

    def __getitem__(self, name):
        try:
            return self.nodes[name]
        except KeyError:
            return None

    def __add__(self, other):
        combined_name = self.name + '_+_' + other.name
        new_graph = graph(combined_name)
        new_graph.nodes = self.nodes
        for nd_name, node_object in other.nodes.items():
            if (nd_name not in new_graph.nodes.keys()):
                new_graph.nodes[nd_name] = node_object
                # print('new_node: {}: {}'.format(nd,new_graph.nodes[nd]))
                # print('\n')
            else:
                for edg, wgh in node_object.neighbors.items():
                    if (edg) in new_graph.nodes[nd_name].neighbors.keys():
                        new_graph.nodes[nd_name].neighbors[edg] = max(new_graph.nodes[nd_name][edg].neighbors, other.nodes[nd_name].neighbors[edg])
                    # print('node {}: neighbor:{}, old weight: {}, new weight:{}'.format(nd,new_graph.nodes[nd][edg], wgh, wgh))
                    else:
                        new_graph.nodes[nd_name].neighbors[edg] = other.nodes[nd_name].neighbors[edg]
                    # print('new neighbor to node {}: {} with weight {}'.format(nd,new_graph.nodes[nd][edg],new_graph.nodes[nd][edg]))
        return new_graph

    def update(self, node):
        if (node.name in self.nodes.keys()):
            for edg, wgh in node.neighbors.items():
                if (edg) in self.nodes[node.name].neighbors.keys():
                    self.nodes[node.name].neighbors[edg] = max(self.nodes[node.name].neighbors[edg], wgh)
                # print('node {}: neighbor:{}, old weight: {}, new weight:{}'.format(nd,new_graph.nodes[nd][edg], wgh, wgh))
                else:
                    self.nodes[node.name].neighbors[edg] = wgh
                # print('new neighbor to node {}: {} with weight {}'.format(nd,new_graph.nodes[nd][edg],new_graph.nodes[nd][edg]))
        else:
            self.nodes[node.name] = node

    def remove_node(self, name):
        if (self.__contains__(name)):
            del self.nodes[name]
        else:
            print(self.name, " does not contain a node named ", name)

    def is_edge(self, frm_name, to_name):
        try:
            return True if to_name in self.nodes[frm_name].neighbors.keys() else False
        except KeyError:
            print('{} is not a node in graph {}'.format(frm_name),self.name)
            return None

    def add_edge(self, frm_name, to_name, weight):
        if (frm_name in self.nodes.keys()):
            if (to_name in self.nodes.keys()):
                if (to_name in self.nodes[frm_name].neighbors.keys()):
                    self.nodes[frm_name].neighbors[to_name] = max(self.nodes[frm_name].neighbors[to_name], weight)
                    # print('node {}: neighbor:{}, old weight: {}, new weight:{}'.format(nd,new_graph.nodes[nd][edg], wgh, wgh))
                else:
                    self.nodes[frm_name].neighbors[to_name] = weight
                    # print('new neighbor to node {}: {} with weight {}'.format(nd,new_graph.nodes[nd][edg],new_graph.nodes[nd][edg]))
            else:
                print('node {} is not in this graph.'.format(to_name))
        else:
            print('node {} is not in this graph.'.format(frm_name))

    def remove_edge(self, frm_name, to_name):
        if (self.is_edge(frm_name, to_name)):
            del self.nodes[frm_name].neighbors[to_name]
        else:
            print('Graph: {} | node: {} | edge: {} | doesn\'t exist'.format(self.name, frm_name, to_name))

    def get_edge_weight(self, frm_name, to_name):
        if (self.is_edge(frm_name, to_name)):
            return self.nodes[frm_name][to_name]
        else:
            return None

    def get_path_weight(self, path):
        if (len(path) == 0):
            print("empty path")
            return None
        elif (len(path) == 1):
            print("Path too short. Length should be at least 2")
            return None
        else:
            for step in path:
                if (step not in self.nodes.keys()):
#print('node {} doesn\'t exist in Graph {}'.format(step, self.name))
                    return None
        total_weight = 0
        for step in range(0, len(path) - 1):
            if (self.is_edge(path[step], path[step + 1])):
                total_weight += self.get_edge_weight(path[step], path[step + 1])
            else:
                print('Path not feasible')
                return None
        return total_weight


    def is_reachable(self, frm_name, to_name):
        stack      = []
        discovered = []
        stack.append(frm_name)
        while (stack):
            current_node = stack.pop()
            if current_node == to_name:
                return True
            if (current_node not in discovered):
                if (current_node == to_name):
                    return True
                discovered.append(current_node)
                
                for connected_node in self.nodes[current_node].neighbors.keys():
                    try:
                        stack.append(connected_node)
                    except KeyError:
                        return False                
        return False

    def get_shortest_path(self, frm_name, to_name):
          shortest_paths = {frm_name: (None, 0)}
          current_node = frm_name
          visited = set()
          while (current_node != to_name):
              visited.add(current_node)
              destinations = self.nodes[current_node].neighbors.keys()
              weight_to_current_node = shortest_paths[current_node][1]
              for next_node in destinations:
                  weight = self.nodes[current_node].neighbors[next_node] + weight_to_current_node
                  if (next_node not in shortest_paths):
                      shortest_paths[next_node] = (current_node, weight)
                  else:
                      current_shortest_weight = shortest_paths[next_node][1]
                      if current_shortest_weight > weight:
                          shortest_paths[next_node] = (current_node, weight)
              next_destinations = {node: shortest_paths[node] for node in shortest_paths if node not in visited}
              if not next_destinations:
                  return "Path not feasible"
              current_node = min(next_destinations, key=lambda k: next_destinations[k][1])
          path = []
          while (current_node is not None):
              path.append(current_node)
              next_node = shortest_paths[current_node][0]
              current_node = next_node
          path = path[::-1]
          return path

        #End of task 1
```

```python
    #Task 3: The roadmap implementation 
        #question 1: From each file create a graph whose nodes are the country regions,
        #and whose edges are the roads themselves (if a travel was not recorded between country regions,
        #then it means such road does not exist). The weight of each edge is defined as the average time
        #(in seconds) of all the travels done on that road. When the two graphs are ready, add them together
        #to create the complete graph of the roadmap.
from datetime import datetime
import time
parts_of_country = ['Center','North','East','West','South']
def graph_from_file(file, graph_name, date_format='%d/%m/%Y %Hh%Mm'):
        g = graph(graph_name)
        denominator_dict = {}
        with open(file, 'r') as my_file:
            countline=0
            for line in my_file:
                if(countline>0 and line.count(',')==3):
                    frm, t1, to, t2=line.split(',')
                    if (frm in parts_of_country and to in parts_of_country):
                        #print(line)
                        t1=t1.strip()
                        t2=t2.strip()
                        try:
                            t1=datetime.strptime(t1,date_format)
                        except ValueError as e:
                            print('Bad departure time format at line {}: {}.'.format(countline,t2))
                            continue
                        try:
                            t2=datetime.strptime(t2,date_format)
                        except ValueError:
                            print('Bad arrival time format at line {}: {}.'.format(countline,t2))
                            continue 
                        if t1 and t2:
                            d1_ts = time.mktime(t1.timetuple())
                            d2_ts = time.mktime(t2.timetuple())
                            minutes_travel = int(d2_ts - d1_ts) / 60
                            if not frm in g.nodes.keys():
                                g.update(node(frm))
                            if not to in g.nodes.keys():
                                g.update(node(to))
                            if not g.is_edge(frm, to):
                                g.add_edge(frm, to, minutes_travel)
                                denominator_dict[(frm,to)] = 1
                            else:
                                prev_average = g.get_edge_weight(frm, to)
                                prev_denominator = denominator_dict[(frm,to)]
                                new_denominator = prev_denominator +1
                                new_average = (prev_average*prev_denominator + minutes_travel)/new_denominator
                                g.add_edge(frm, to, new_average)
                                denominator_dict[(frm, to)] = new_denominator
                                #print('from {} to {} new data: {}, new average: {}'.format(frm,to,minutes_travel,new_average))
                        else:
                            print('Bad departure/arrival time format at line {}: {} , {}.'.format(countline,t1,t2))
                    else:
                        print ('Bad region at line',countline)

                countline +=1
            print ('read', str(countline-1), 'lines')
        return g
g1= graph_from_file('travelsew.csv','travelsew')
g2= graph_from_file('travelswe.csv','travelswe','%I:%M:%S%p ; %b %d %y')
print(g1)
print(g2)
g1.__add__(g2)
print(g1)
print('-------------------------------------------------------------------------\n')
```

```python
    #Task 2:The social network implementation
def social_graph(file, graph_name):
        g = NonDirectionalGraph(graph_name)
        count_friendships=0
        max_friendships=0
        with open(file, 'r') as my_file:
            countline=1
            for line in my_file:
                    person1, person2, status=line.split()[0],line.split()[2],line.split()[3]
                    if(not g.__contains__(person1)):
                        g.nodes[person1]=node(person1)
                    if(not g.__contains__(person2)):
                        g.nodes[person2]=node(person2)
                    if(status=='became'):
                        g.add_edge(person1,person2)
                        #print('{}+{}= Friendship!'.format(person1,person2))
                        count_friendships+=1
                    elif(status=='cancelled'):
                        g.remove_edge(person1,person2)
                        #print('{}+{}= Breakup!'.format(person1,person2))
                        count_friendships-=1
                    else:
                        print('Bad format on line. Unsure of the action.')
                    countline+=1
                    max_friendships=max(max_friendships,count_friendships)
                        
        return g,max_friendships
```
